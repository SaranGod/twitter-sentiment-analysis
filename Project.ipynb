```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from keras.layers import Dense
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
```

```python
data=pd.read_csv("train_utf.csv")
```

```python
data.head()
```

```python
data.SentimentText[3]
```

```python
data.Sentiment.value_counts()
```

```python
train_size=int(len(data)*.8)
```

```python
train_text=data['SentimentText'][:train_size]
train_sent=data['Sentiment'][:train_size]
```

```python
train_sent
```

```python
test_text=data['SentimentText'][train_size:]
test_sent=data['Sentiment'][train_size:]
```

```python
max_words=1500
tokenize=Tokenizer(num_words=max_words,char_level=False)
```

```python
tokenize.fit_on_texts(train_text)
x_train=tokenize.texts_to_matrix(train_text)
x_tests=tokenize.texts_to_matrix(test_text)
```

```python
x_train
```

```python
tokenize.index_word
```

```python
tokenize.word_docs
```

```python
encoder=LabelEncoder()
encoder.fit(train_sent)
```

```python
y_train=encoder.transform(train_sent)
y_test=encoder.transform(test_sent)
```

```python
encoder.classes_
```

```python
y_train
```

```python
import numpy as np
```

```python
np.unique(y_train)
```

```python
from keras import Sequential
```

```python
model=Sequential()
```

```python
model.add(Dense(512,input_shape=(max_words,),activation='relu'))
```

```python
model
```

```python
model.add(Dense(2,activation='softmax'))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
from keras import utils
```

```python
y_train_new=utils.to_categorical(y_train)
y_test_new=utils.to_categorical(y_test)
```

```python
history=model.fit(x_train,y_train_new,batch_size=32,epochs=2,verbose=1,validation_split=0.1)
```

```python
score=model.evaluate(x_tests,y_test_new)
```
